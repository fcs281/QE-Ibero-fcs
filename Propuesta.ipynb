# **Propuesta de factibilidad o metodológica del uso de Quantum Espresso en la Universidad Iberoamericana**. 


Quantum Espresso (QE) es un código de acceso libre para el cálculo de propiedades electrónicas de sistemas de materiales. Su núcleo es la solución de la ecuación de Schrödinger del sistema mediante la Teoría Funcional de la Densidad, el uso de pseudopotenciales, ondas planas y el teorema de Bloch. Códigos como QE han sido un parteaguas en la simulación de materiales, permitiendo avances inimaginables en diferentes campos desde la física hasta la medicina.

Pople, fueron galardonados con el Premio Nobel de Química por *"el desarrollo de la teoría de la funcional de densidad y por el desarrollo de métodos computacionales para química cuántica"* respectivamente. Dada la flexibilidad para obtener información estructural y electrónica de modelos de materiales a diferentes escalas, es una herramienta que considero podría ser de gran utilidad para estudiantes de licenciatura de Ingenierías, específicamente de Ingeniería Física. 


Partiendo de la experiencia de trabajar durante mi doctorado con un código similar, pero no gratuito, [CASTEP](http://www.castep.org/), así como de algunos ejercicios con dicho código en la materia *"Temas selectos de física de materiales"* (Plan 2004), y a la experiencia del profesor Dr. Ángel Reyes Retana utilizando QE en la materia *"Ciencia y tecnología de nanomateriales"* (Plan SUJ) con resultados positivos, decidí aprender el codigo y proponer su uso para la realización de proyectos en la carrera de Ingeniería Física. 

Al momento de hacer la propuesta de entregables para mi sabático, propuse un estudio de factibilidad para implementar el código en la Ibero, sin embargo, al usarlo, me dí cuenta que por el avance en hardware y software facilidad de instalar y "correr" el código en PCs o Laptos convencionales. Para uso didáctico sin fines de investigación, no es necesario disponer supercómputo, clusters, servidores o similare. Por lo tanto, el presente trabajo tendra el objetivo de:

    
**Proporcionar una guía para la instalación y uso académico de Quantum Espresso para estudiantes de Ingeniería Física, apoyado del uso de Python para desarrollar habilidades de programación y de manejo de datos, de la mano de los conocimientos de física del estado sólido que se requieren para comprender la metodología y analizar los resultados de los cálculos**.
