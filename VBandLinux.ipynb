## Generar una Máquina Virtual en Windows con una distribución de Linux

El sistema operativo ideal para instalar y correr QE es Linux, sin embargo actualmente el código corre en la mayoría de las distribuciones de Linux, Mac-OS y Windows, y se puede encontrar en la página oficial de la organización: [Download-QE](https://www.quantum-espresso.org/download). 
Dado que la mayoría de los usuarios a los que va dirigido esta guía son usuarios de Windows, en este notebook se cubrirán los pasos para realizar una instalación de un [Virtual Box de Oracle](https://www.virtualbox.org/) en el sistema Windows, generar una máquina virtual adecuada para finalmente instalar una distribución [Debian](https://www.debian.org/index.es.html) de Linux en donde posteriormente se instalará QE. Además en otro notebook bajaremos Python para utilizar algunos programas en Python para trabajar con QE [QE-doc](https://github.com/jochym/qe-doc).


 ## 2.1 Instalar el administrador de Oracle VM Virtual Box
 
Mediante el programa llamado **Oracle VM VirtualBox Administrator**, Oracle ofrece la posibilidad de instalar diferentes máquinas (computadoras) virtuales de diferentes sistemas operativos -huespedes-, en cualquier sistema operativo -anfitrión-. Aquí, instalaremos en el anfitrión Windows, el programa de Oracle para poder instalar uno o más huespedes de Linux, específicamente la distribución Debian.

Para esto, seguiremos los siguientes pasos:

* De la página de [Oracle](https://www.virtualbox.org/) bajar en "download" la versión para windows (VirtualBox-6.0.10-13207-Win.exe o similar, la versión probada mas actualizada, yo trabajé con la 5.2.20).
* Dar doble click en el ejecutable para su instalación.
* Click "next" para comenzar la instalación.
* Aceptar la localización dentro de Windows en donde se instalará la aplicación "VirtualBox", o destinar una ubicación específica.  En un paso siguiente, instalaremos en esta ubicación un sistema operativo Linux, por lo que **es importante asegurarse de que el disco en donde se instalará la aplicación tiene capacidad de almacenamiento suficiente para destinar algunas decenas de GB al nuevo sistema operativo, sin que afecte al sistema original (windows)**. 
* Seleccionar la manera en que se instalarán los ejecutables.
* Aceptar una desconección temporal de la red.
* De ser necesario, revisar los parámetros de instalación, si no es necesario: "install".
* Durante la instalación, instalarlos controladores (confiar en Oracle Corporation).
* Finalizar e iniciar "Oracle VM VirtualBox".

Un documento pdf incluyendo la impresión de pantalla de todos los pasos anteriormente mencionados se puede encontrar en el siguiente vínculo: [tutorial instalación VB-Oracle.pdf](Docs/VB-Oracle.pdf)

De este modo quedará instalado un administrador de máquinas virtuales, en donde se podrán crear máquinas para instalar los sistemas operativos necesarios.

## 2.2 Configurar la máquina virtual para Debian
 
Para instalar el nuevo sistema operativo, es necesario generar una **máquina virtual** o **virtual box** que albergará al mismo.
### Descargar Debian
Previo a la instalación se requiere:
1. Contar con la imagen de la distribución de Debian que se va a instalar. De preferencia guardar dicha distribución en una memoria USB con un mínimo de memoria de 500 MB. Una imagen pequeña descargable se puede encontrar en la siguiente figura de la página web: [debian.org/distrib/netinst](debian.org/distrib/netinst). 
De entre las opciones, las marcadas en naranja de la figura siguiente, son las recomendadas. Elegir ya sea la versión **amd64** o **i386**, dependiendo del sistema operativo anfitrión, si es de 64- o 32-bits respectivamente.
2. Contar con al menos 30 GB de almacenamiento libre en una unidad de disco duro de la maquina anfitriona.

### Generación/configuración de Máquina Virtual (Virtual Box VB)
La siguiente lista indica los pasos para la creación de la máquina virtual (una impresión de pantalla de cada paso está disponible en el archivo [(Generación de Máquina Virtual para Debian)](/Docs/VB-Debian.pdf )(*dependiendo de la versión o de las selecciones que se vayan haciendo en el camino, puede haber algunas diferencias).

Se trabajará directamente en la ventana de [Oracle VM VirtualBox Administrator](Figs/VB-Oracle.png).

* En la aplicación "Oracle Virtual Box Administrator" seleccionar "nueva"
* Definir el sistema operativo: elegir Linux y Debian (32 o 64 dependiendo del sistema operativo anfitrión).
    * Elegir un nombre para la nueva máquina virtual, *e.g.*: LinuxFelipePCIbero.
    * Definir una carpeta para que se instale el sistema operativo Linux. -Preferentemente aceptar la sugerencia del instalador para que esté en la ubicación de VB-.
    * Elegir en Tipo: *Linux.*
    * En Versión: *Debian.*
    * Seleccionar "modo experto" para asegurar que el "la cantidad de memoria" destinada no comprometa al sistema windows. *E.g.* si la máquina tiene 8 GB de memoria, elegir el tamaño de la memoria de la mitad, siempre en el rango verde de la escala disponible. 
    * Seleccionar: Crear.
    * En modo experto, definir el tamaño del archivo, asegurarse de que no se compromete ni al sistema Windows, ni se sacrifique el tamaño del nuevo sistema, por ejemplo, en un disco duro de 500 GB con 150 GB libres, destinar 50 GB al nuevo sistema Linux.  En este punto es importante **revisar el espacio libre en windows**.
    * Seleccionar el tipo de archivo de disco duro como VDI y el Almacenamiento en unidad de disco duro física Reservado dinámicamente.
    * Seleccionar: Crear. 
* El administrador de Oracle ha creado una máquina virtual adecuada para instalar un sistema operativo Debian-Linux, **¡¡Felicidades!!**.

## 2.3 Instalación de Debian. 

Al igual que en las secciones anteriores, el siguiente archivo contiene impresiones de pantalla de los pasos para instalar Debian, mismos que, dependiendo de cada máquina y-o versión del sistema operativo, o selección personal de alguna opción, pueden cambiar. Archivo [Instalación de Debian](Docs/DebianInstall.pdf):


* En la pantalla de Oracle VM VirtualBox Administrator, selecionar el nuevo sistema operativo.
* Conectar a la computadora la USB con la imagen de Debian.
* Asegurarse de contar con una buena conexión a internet. 
* Seleccionar *"Almacenamiento"*, para definir desde dónde se descargará la imagen de Debian.
* Seleccionar en "Controller:IDE", "Empty", después de lo que se desplegarán atributos del drive óptico de la computadora. 
* En "*Attributes*", "*Optical Drive*", seleccionar cualquier IDE. 
* Dar click a la lupa de la izquierda y seleccionar cualquier unidad. Esto desplegará un buscador para
* Buscar la unidad USB y seleccionar la imágen de Debian anteriormente descargada.
* Dar click en OK. Con esto, el administrador de Virtual Box, tiene la información de dónde leer la Imagen de Debian.
* Dar clik en Iniciar.
* Se desplegará una pantalla de Instalación en donde hay que seleccionar "*Graphical install*" y Enter.
* Seleccionar el idioma, localización de su preferencia. En el archivo se selecciona México.
* Seleccionar la combinación de idioma y localización.
* Indicar el idioma del teclado de la máquina. Preferentemente con el que esté configurada la máquina en Windows, o el que venga de fábrica con la computadora.
*  A continuación pueden despelgarse diferentes pantallas, pero llevarán hasta la configuración de el "*hostname*" del sistema operativo, *E.g.* FelipeTestDebian
* Indicar el dominio, *E.g.* fcs281.
* Escribir dos veces la contraseña del usuario "*root*". *E.g.* password.
* Definir al usuario y contraseña. *E.g.* felipe y userpassword.
* Configurar el reloj.
* Configurar la partición del disco. Preferntemente seleccionar "*Guided-use entire disk*".
* Para la partición del disco, seleccionar la opción única. Continuar.
* El siguiente paso indica la partición de los discos, dependiendo de las preferencias realizar la selección. Para usuarios primerizos, se recomienda *All files in one partition*.
* Finalizar la partición, aceptar los cambios al disco.
* Se instalará la base del sistema, tardando algunos minutos. 
* Para la configuración del "*package manager*", indicar que no se cuenta con otro DVD o CD (esto llevará a la selección de un repositorio o "*mirror*", desde donde se terminarán de bajar archivos para el sistema).
* Para la configuración de los paquetes, seleccionar a su gusto entre las opciones. Para este tutorial se probaron tanto United States como México, ambas funcionan bien.
* El "*Debian archive mirror*" recomendado es **ftp.us.debian.org**, sin embargo *mirrors* de diferentes uniersidades funcionan bien.
* La pantalla donde pregunta por el "*HTTP proxy*", dejarla en blanco y continuar.
* Se seguirá instalando el sistema. Puede tardar unos minutos.
* Seleccionar si se desea o no participar en envío de datos del sistema operativo. 
* La pantalla de "*Software selection*", permite preinstalar software, se recomienda elejir un escritorio, puede ser el de Debian, o si se cuenta con espacio suficiente, GNOME es buena opción. Las opciones de abajo, preseleccionadas, dejarlas como están. 
* Se continuará instalando software. 
* Seleccionar que se instalen actualizaciones de seguridad de manera automática.
* Indicar, si, para el *GRUB boot loader*.
*  Seleccionar la partición indicada para el *GRUB boot loader*, no de manera manual.
* ** Finalmente aceptar que se ha instalado el programa y reiniciar el sistema.**
* Antes de que se reinicie el sistema, remover la USB.
* Ahora aparecerán las típicas inicializaciones de Linux, y eventualmente aparecerá la pantalla de inicio con el nombre del usuario anteriormente definido. Poner el password y disfrutar de la nueva **computadora virtual con sistema operativo Linux, montada en la máquina Windows.**

## 2.4 Ya en Linux...

Linux es un sistema operativo muy versatil. Para poder trabajar con Quantum Espresso es necesario instalar una serie de compiladores y tener los ejecutables. Es necesario contar específicamente con:
    * gcc
    * g++
    * gfortran
    * make  
 
Dar click en Activities y teclear "*terminal*"
Se abrirá una nueva ventana llamada "*terminal*" desde donde uno puede realizar toda serie de tareas. Aquí uno puede bajar los programas o paquetes que hagan falta. 
* En la terminal, entar como superusuario, es decir, teclear "su".
* Dar la contraseña que se le dió al administrador, en mi caso "password".
* teclear *sudo apt-get update*
* teclear *sudo apt-get install "gcc"*
* La terminal pedirá permiso para disponer de espacio de disco, indicar "*Y*".
* De este modo se pueden instalar todos los paquetes o software que hacen falta. (Recordar que hay que estar conectado a internet para que Linux busque en el "*mirror*" la información y pueda realizar las instalaciones.

El próximo paso es instalar el programa "*Open MPI*" que permitirá que Quantum Espresso realice cálculos en paralelo:

        
### Instalar "Open MPI"
* Instalar el "open mpi" de [OpenMPI](http://www.open-mpi.org/), [v4.0](https://www.open-mpi.org/software/ompi/v4.0/),[ayuda](https://www.open-mpi.org/faq/?category=rsh), [guía de instalación](http://lsi.ugr.es/jmantas/pdp/ayuda/instalacion.php?ins=openmpi):
* Instalarla directamente en "root", no en "home", "usr", "programs", etc. 
* *cd /*  (para ir a root)
* *tar xzvf openmpi-version.tar.gz*  (para extraer la info del *.tar.gz)      
* *su* (para ingresar como superusuario)
* *./configure -prefix=”/home/user/.openmpi”* Crea una sere de carpetas
* *make; sudo make install*
* Para que la exportación de variables perdure para las siguientes sesiones y terminales, apuntaremos los comandos siguientes en el fichero de variables de entorno. Por defecto será "/home/\<usuario\>/.bashrc" para los usuarios de bash.
* *echo export PATH="\$PATH:/home/\$USER/.openmpi/bin" >> /home/\$USER/.bashrc ;*
* *echo export LD\_LIBRARY\_PATH="\$LD\_LIBRARY\_PATH:/home/\$USER/.openmpi/lib/" >> /home/\$USER/.bashrc*
* probar que mpirun está instalado.
