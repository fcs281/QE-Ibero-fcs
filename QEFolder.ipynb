# 4. ¿Qué contiene la carpeta instalada de QE-6.3?

Al instalar QE se crea una carpeta , *e.g.* "qe-6.3", en donde se almacena toda la información  del programa: "file sources", documentación, herramientas y otros programas. A continuación se presenta una lista de la mayoría de los elementos que aparecen en dicha carpeta, con una brevísima descripción de los mismos, los elementos que utilizaremos para ejemplificar el uso de QE se marcan en **negritas**:

* XSPECTRA: Código para "X-ray spectra calculation".
* test-suite: Contiene una serie de ejecutables para pruebas de cálculos o *benchmarks*, con una descripción de lo que se prueba. 
* S3DE: "A suite of scripts used to manage an environment suitable for code development. Automatic Makefile generation, versions, portability among different systems and library loading are some of the main tasks of the DE".
* QHA: Código "Quasi-Harmonic Approximation". Calcula fonones y proyecciones de las DOS fonónicas.
* PWCOND: Calcula estructura de bandas compleja, kpoints complejos, potenciales autoconsistentes, transmitancia, etc.
* **PW: "Plane-Wave Self-Consistent Field", contiene un "overview", ejemplos, documentación, código y herramientas ("A collection of simple tools for specific tasks", e.g. "pwi2xsf.sh, pwo2xsf.sh : convert a pw.x input or output to the xcrysden format").**
* **PP:"PostProc", "an open-source package for postprocessing of data produced by \PWscf\ and \CP." E.g.: "The main postprocessing code \texttt{pp.x} extracts the specified data from the data files produced by \PWscf\ (\pwx\ executable) or \CP\ (\cpx\ executable); prepares data for plotting by writing them into formats that can be read by several plotting programs."**
* PlotPhon: Programa para graficar densidad de estados de fonones. 
* PHonon: información y código necesario para realizar cálculos de fonones o de fonones en el punto gamma. 
* NEB: "open-source package for the calculation of energy barriers  and reaction pathway using the Nudged Elastic Band (NEB) method"
* License: información y licencia del programa.
* KS_Solvers: Rutinas que resuelven la ecuación de "Kohn Sham".
* GWW: GW quasi-particle calculations, use of the "Bethe–Salpeter equation" for two body particles quantum fields theoretical calcs. 
* GUI: Stands for "Graphical-user-interfaces". The folder contains:
    *(1) [PWgui](http://www-k3.ijs.si/kokalj/pwgui/pwgui.html): a Graphical-User-Interface for PWscf, [CompMatSci2003](https://www.sciencedirect.com/science/article/pii/S0927025603001046), [PDF](Kokalj_Computer_graphics_and_graphical_user_interfaces_as_tools_in_simulations_of_matter_at_the_atomic_scale_CompuMaterSci_2003.pdf)
    *(2) QE-modes package: a collection of Emacs major-modes for Quantum ESPRESSO input files."
* EPW:" "Electron-phonon Wannier". EPW is an open-source F90/MPI code which calculates properties related to the electron-phonon interaction using [Density-Functional Perturbation Theory](DFPT-DalCorso-Shanghai_2008.pdf) and Maximally Localized Wannier Functions".[EPW](http://epw.org.uk/)
* environment_variables: Contiene las variables ambientales del sistema, e.g. "BIN_DIR = path of compiled executables. Usually this is \$PREFIX/bin, where \$PREFIX is the root of the Quantum ESPRESSO source tree."
* **Doc: Contiene la documentación general de QE. Entre otros documentos contiene: user-guide.pdf, developer-man.pdf, plumed_quick_ref.pdf y brillouin_zone.pdf.**. Estos documentos pueden ser de utilidad para la instalación o manejo del programa. 
* dev-tools: Opciones avanzadas para desarrolladores.
* CPV: "Car-Parrinello molecular dynamics code". Contiene documentación y código de dinámica molecular, incluyendo dinámica de celda variable y cálculo de superficies de energía libre. [P. Giannozzi, O et al. J.Phys.:Condens.Matter 29, 465901 (2017)](https://iopscience.iop.org/article/10.1088/1361-648X/aa8f79)
* COUPLE: "This package allows calling programs of the QE package as subroutines"
* CONTRIBUTING.md: Archivo que explica las modalidades de posibles contribuciones en el código QE.
* atomic: Código para cálculos de estructura electrónica y generación de pseudopotenciales. Contiene guías de la teoría de PP, así como para la generación y pruebas de los PPs generados. 
* archive: Carpeta que contiene los archivos .tar.gz de librerías y archivos necesarios. 
* include: Archivos de configuración usados y generados durante la configuración, compilación del paquete. 
* clib: codigo.
* FoX: archivos binarios y librerías del código.
* UtilXlib: archivod de código.
* LAPACK -> lapack-3.6.1: "a library of Fortran 90 with subroutines for solving the most commonly occurring problems in numerical linear algebra." [Lapack](http://www.netlib.org/lapack/)
* LAXlib: Carpeta que contiene librerías para de prueba y de cronometraje ("timing code"). Pruebas de comunicación entre procesadores, número de procesadores, tareas específicas, etc. 
* FFTXlib: "Implements real space grid parallelization of FFT and task groups. " También contiene archivos para realizar pruebas de evaluación de paralelización. Incluye un "python script to extract the parameters from an output file of pw.x".
* Modules: Código. 
* dft-d3: "Interface routines for Quantum ESPRESSO are contained in file dftd3_qe.f90.This file is released under the GPL."
* install: Archivos de instalación. 
* LR_Modules: Archivos de código
* upftools: Herramienta automática para generar pseudopotenciales. También tiene herramientas para convertir entre pseudopotenciales para QE y CASINO y viceversa. 
* TDDFPT: Carpeta con codigo para realizar cálculos "Time Dependent Density Functional Perturbation Theory (TDDFPT)"
* iotk: redirecciona a subfolder de *S3DE* mencionado arriba. 
* pseudo: Carpeta a donde se bajan y almacenan los pseudopotenciales.
* bin: Carpeta que contiene ejecutables de código y de rutinas de análisis y manejo de postprocesamiento.
* tempdir: Aquí se almacenan archivos temporales. Si no se redefine el tempdir al realizar un cálculo, los archivos de las funciones de onda son guardados en esta carpeta.
