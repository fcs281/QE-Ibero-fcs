# 3. Instalar Quantum Espresso (QE)

A continuación procederemos a la instalación del programa QE, basándonos en parte en las instrucciones publicadas en [www.fisica.uniud](http://www.fisica.uniud.it/~giannozz/QE-Tutorial/installation.html):
* Se recomienda no instalar como "superusuario" (su), e intalar en la carpeta "*root.*" 
* Bajar versión estable del programa para Linux de la página de [github QE](https://github.com/QEF/q-e/tags). Especificamente se bajará un archivo *nombre-version.tar.gz*, *e.g.*. qe-6.3.tar.gz.
* Utilizando la terminal, en *root* o la ubicación donde se vaya a instalar y en donde se encuentre el programa teclear: *tar xvzf qe-6.3.tar.gz*
* se creará la carpeta: qe-6.3. Posteriormente se puede borrar el archivo .tar.gz.
* En la carpeta de QE :
    * Teclear: *./configure*
    * Continuar con: *make all*
    * Ahora: export *LD\_LIBRARY\_PATH=/usr/local/lib/:\$LD\_LIBRARY\_PATH* 
        * también funciona: *export LD\_LIBRARY-PATH=\$LD\_LIBRARY\_PATH:/usr/local/lib*
