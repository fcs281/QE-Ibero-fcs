## Editores de código

Podríamos usar Python directamente desde la consola de comandos de nuestra computadora, sin embargo lo usual es utilizar un entorno de desarrollo integrado o "integrated development environment" (IDE) con una interfaz gráfica más amigable que nos permitirá editar el código y además contienen funcionalidades adicionales que hacen más fácil la tarea de programar. Algunas de los editores más usadas son [Jupyter Notebooks](http://jupyter.org), [Spyder](https://github.com/spyder-ide/spyder), [Atom](https://atom.io) y [PyCharm](https://www.jetbrains.com/pycharm/).

En esta presentación vamos a utilizar el Jupyter notebook ya que es más amigable para empezar a programar porque nos permite de ver de manera interactiva y línea a línea los resultados de lo que vamos programando.("_Jupyter Notebook is an open source web app that allows developers to create and maintain notebook documents. ## Abriendo Jupyter Notebook
 Con anaconda se instalan dos programas con los que podemos trabajar Python: **Spyder** y **Jupyter Notebook**. 

Los notebooks nos permiten integrar código, texto markdown, visualizaciones y otros recursos multimedia en un solo documento.

Para abrir Jupyter Notebook tenemos dos opciones:
1. Buscar entre nuestros programas instalados a Jupyter Notebook y abrirlo.
2. Desde la terminal de comandos, nos ubicamos en la carpeta en que queremos trabajar y escribimos: 
```
jupyter notebook
```

Se abrirá una ventana de comandos que **no debemos cerrar**.

En nuestro explorador de internet se abrirá una nueva pestaña que será el host local de Jupyter (por defecto es http://localhost:8888/tree? ), y mostrará los archivos de ese directorio.

### Creando un nuevo notebook

1. En la pestaña Home que se abre, nos vamos a la esquina superior derecha, hacemos click en el botón que dice *New* y seleccionamos *Python [root]*

2.  Se abrirá una nueva pestaña con un notebook y le asignamos un nombre. 
3.  En los notebooks se trabaja con celdas, en las que puede ir código o texto en markdown (permite incluir expresiones matemáticas de latex $y = X^2\hat{\beta} + \mu_i$). En la caja superior se puede seleccionar cuál es el contenido de la celda.

4. Para agregar nuevas celdas hacemos click en el el símbolo $+$ en la parte superior.

5. Para ejecutar el código de una celda presionamos *ctrl + enter*. El resultado de la operación, si lo hay, se mostrará debajo de la celda.
